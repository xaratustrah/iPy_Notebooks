### Convolution plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib osx 
```

```python
def conv_plot(a, b, c):
    data = np.array([a, b, c])

    ax = mkl = stl = bl = [None] *len(data)

    fig1 = plt.figure()
    for i in range(len(ax)):
        ax[i] = plt.subplot(3,1,i+1)
        ax[i].set_xlim([0,max(len(a), len(b), len(c))+1])
        ax[i].set_ylim([-1,max(max(a), max(b), max(c))+1])        
        mkl[i], stl[i], bl[i] = ax[i].stem(data[i], '-.')
        plt.setp(mkl[i], 'markerfacecolor', 'b')
        plt.setp(bl[i], 'color', 'r', 'linewidth', 2)    
        plt.grid()
```

```python
a = np.random.randint(0, high=3, size=20)
a
```

```python
b= np.array([2,1,2])
brev = np.fliplr([b])[0] # trick to make it work with 1D array
conv_plot(a, b, np.convolve(a, b))
```
