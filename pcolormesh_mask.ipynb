```python
import matplotlib.pyplot as plt
import numpy as np
from copy import copy
%matplotlib inline
```

```python
xx, yy = np.meshgrid(np.arange(64), np.arange(32))
zz = np.random.random(size=(32, 64))
```

```python
plt.pcolormesh(xx, yy, zz, cmap=plt.cm.jet)
```

```python
palette = copy(plt.cm.jet)
palette.set_bad('white')
```

```python
zz = np.ma.masked_where(zz < 0.2, zz)
#zz[zz<0.2] = np.NaN
```

```python
plt.pcolormesh(xx, yy, zz, cmap=palette)
```
