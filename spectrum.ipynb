# Spectrum calculation and correlation of Schottky signals

```python
import numpy as np
from iqtools import *
from pylab import specgram, cm, psd
import matplotlib.pyplot as plt
from cmath import phase
%matplotlib inline
```

```python
filename1 = './decays/RSA51-2014.10.02.08.55.23.072.TIQ'
dic1,_ = read_tiq(filename1, 1500)
x1 = dic1['data']
fs1 = dic1['fs']
lframes1 = dic1['lframes']
center1 = dic1['center']
nframes_tot1= dic1['nframes_tot']
```

```python
filename2 = './decays/RSA51-2014.10.02.08.56.44.571.TIQ'
dic2,_ = read_tiq(filename2, 100, 1024, 1500)
x2 = dic2['data']
fs2 = dic2['fs']
lframes2 = dic2['lframes']
center2 = dic2['center']
nframes_tot2= dic2['nframes_tot']
```

```python
f1, y1 = plot_fft(x1, fs1, 0)
f2, y2 = plot_fft(x2, fs2, 0)
```

```python
Pxx1, freqs1 = psd(x1, NFFT=lframes1, Fs = fs1, noverlap=512)
Pxx2, freqs2 = psd(x2, NFFT=lframes2, Fs = fs2, noverlap=512)
```

```python
Pxx1, freqs1, bins1, im1 = specgram(x1, NFFT=lframes1, Fs=fs1, noverlap=512, cmap=cm.binary_r)
```

```python
Pxx2, freqs2, bins2, im2 = specgram(x2, NFFT=lframes2, Fs=fs2, noverlap=512, cmap=cm.gist_heat)
```

```python
plt.plot(x1.real[0:100], x1.imag[0:100], 'r-')
plt.plot(x2.real[0:100], x2.imag[0:100], 'g-')
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
```

```python
plt.plot(np.angle(x1[0:100], deg=True), 'r')
plt.plot(np.angle(x2[0:1000], deg=True), 'g')
plt.ylabel("Phase [deg]")
```
