## Test fortran magic

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import random
%matplotlib inline
%load_ext fortranmagic
```

```python
a = np.array([random.random() for s in range(10)])
```

```python
a
```

```python
%%fortran
    subroutine my_function(x, y, z)
        real, intent(in) :: x(:), y(:)
        real, intent(out) :: z(size(x))
        ! using vector operations
        z(:) = sin(x(:) + y(:))
    end subroutine
```

```python
plt.plot(my_function(a, a**2), 'b-o')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('Fortran test')
```
