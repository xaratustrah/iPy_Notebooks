# Relation of Venus and Earth orvbit

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from iqtools import *
from scipy import signal
from mpl_toolkits.mplot3d import Axes3D
import time
```

```python
tv, v = make_signal(6.15, 1000, l = 10, noise=False)
te, e = make_signal(10, 1000, l = 10, noise=False)
```

```python
plt.plot(v,e, color='r')
plt.grid(1)
plt.savefig('earth_venus.pdf')
```

```python
xbar_v, instph_v = make_analytical(v)
xbar_e, instph_e = make_analytical(e)
```

```python
plt.gca().set_aspect('equal')
plt.grid(False)
plt.axis('off')
for i in range (1, len(xbar_e), 200):
    plt.plot([np.real(xbar_e)[i], np.real(xbar_v)[i]], [np.imag(xbar_e)[i], np.imag(xbar_v)[i]], 'r')
plt.savefig('earth_venus.pdf')
```
