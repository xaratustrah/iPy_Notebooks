### date time in matplotlib

```python
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
```

```python
others = []
dates = []

with open('gasjet_neu.dat') as f:
    for row in f:
        array = f.readline().split()
        dates.append(dt.datetime.strptime(array[0], '%Y-%m-%d_%H:%M:%S'))
        others.append(array[2:])
othersnp = np.asarray(others)
```

```python
fig, ax = plt.subplots(1)
fig.autofmt_xdate()
plt.plot(dates, othersnp[:,0])
xfmt = mdates.DateFormatter('%Y-%m-%d %H:%M')
ax.xaxis.set_major_formatter(xfmt)
plt.grid()
```
