# Cassini's ovals

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theta = np.arange(2*np.pi/3,6*np.pi,0.01)
theta = np.arange(-np.pi/2,np.pi/2,0.01)
def f1(theta):
    return a**2*np.cos(2*theta)+np.sqrt((np.power(a,4)*np.cos(2*theta)**2+np.power(k,4)-np.power(a, 4)))
def f2(theta):
    return a**2*np.cos(2*theta)-np.sqrt((np.power(a,4)*np.cos(2*theta)**2+np.power(k,4)-np.power(a, 4)))
```

```python
a = 0.5
k = 0.4
plt.polar(theta, f1(theta), lw=2)
plt.polar(theta, f2(theta), lw=2) 
```
